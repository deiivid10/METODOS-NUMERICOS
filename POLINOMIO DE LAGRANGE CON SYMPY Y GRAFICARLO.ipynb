# CALCULAR EL POLINOMIO DE LAGRANGE CON SYMPY Y GRAFICARLO

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

puntos = [(1, 2), (2, 3), (3, 5)]
x = sp.Symbol('x')
print("Puntos de Interpolacion: ", puntos)
def polinomio_lagrange(puntos):
    n = len(puntos)
    polinomio = 0
    for i in range(n):
        xi, yi = puntos[i]
        termino = yi
        for j in range(n):
            if i != j:
                xj, _ = puntos[j]
                termino *= (x - xj) / (xi - xj)
        polinomio += termino
    return sp.simplify(polinomio)

polinomio = polinomio_lagrange(puntos)
print(polinomio)
polinomio_func = sp.lambdify(x, polinomio, 'numpy')
x_val = np.linspace(min(p[0] for p in puntos) -1 , max(p[0] for p in puntos) + 1, 100)
y_val = polinomio_func(x_val)
plt.plot(x_val, y_val, label='Polinomio de Lagrange')
plt.scatter(*zip(*puntos), color='red', label='Puntos de interpolación')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.legend()
plt.grid()
plt.show()



```
