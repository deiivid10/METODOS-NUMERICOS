```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = sp.ln(x)
```

# Definir la variable y la función ⬆


```python
x0 = 1
taylor_series = sp.series(f, x, x0, n=5).removeO() # n=5 porque queremos hasta (x-x0)^4
```

# Obtener la serie de Taylor de cuarto orden en x0 = 1 ⬆


```python
taylor_func = sp.lambdify(x, taylor_series, 'numpy')
```

# Convertir la serie en función numérica ⬆


```python
x_vals = np.linspace(0.2, 2, 400) # Evitamos x=0 porque ln(0) no está definido
f_vals = np.log(x_vals) # Valores reales de ln(x)
taylor_vals = taylor_func(x_vals) # Valores del polinomio de Taylor
```

# Crear valores para la grafica  ⬆

```python
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_vals, label=r'$f(x) = \ln(x)$', color='blue')
plt.plot(x_vals, taylor_vals, label=r'Taylor (4° orden)', linestyle='--', color='red')

plt.scatter([x0], [np.log(x0)], color='black', zorder=3, label=r'$x_0 = 1$')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Aproximación de Taylor de ln(x)')
plt.legend()
plt.grid(True)
```
