```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función que se va a integrar: f(x) = e^x
def f(x):
    return np.exp(x)

# Implementación de la Regla de Simpson Compuesta (Algoritmo 4.1)
def simpson_compuesta(f, a, b, n):
    if n % 2 != 0:
        raise ValueError("n debe ser par")  # La regla de Simpson requiere que n sea par
    h = (b - a) / n  # Longitud de cada subintervalo
    x = np.linspace(a, b, n + 1)  # Se generan los n+1 puntos entre a y b
    fx = f(x)  # Se evalúa la función en todos esos puntos
    # Se aplica la fórmula de Simpson compuesta:
    # f(a) + f(b) + 4*f(puntos impares) + 2*f(puntos pares internos)
    S = fx[0] + fx[-1] + 4 * sum(fx[1:-1:2]) + 2 * sum(fx[2:-2:2])
    return (h / 3) * S  # Resultado final de la integral aproximada

# Parámetros del intervalo de integración
a = 0  # Límite inferior
b = 4  # Límite superior
n = 2  # Número de subintervalos (debe ser par)

# Cálculo de la integral aproximada usando Simpson compuesta
aprox = simpson_compuesta(f, a, b, n)

# Cálculo del valor exacto de la integral definida de e^x en [0,2]
exacto = np.exp(2)

# Cálculo del error absoluto entre el valor exacto y el aproximado
error = abs(exacto - aprox)

# Impresión de resultados
print(f"Aproximación con n={n}: {aprox}")
print(f"Valor exacto (exp(2)): {exacto}")
print(f"Error absoluto: {error}")

# Generación de valores para graficar la función original
x_vals = np.linspace(a, b, 1000)  # Muchos puntos para una curva suave
y_vals = f(x_vals)  # Evaluación de f(x) = e^x en esos puntos

# Puntos que se usan en la Regla de Simpson
x_simpson = np.linspace(a, b, n + 1)  # Puntos de evaluación
y_simpson = f(x_simpson)  # f(x) evaluado en esos puntos

# Crear la gráfica
plt.figure(figsize=(8, 5))  # Tamaño de la figura
plt.plot(x_vals, y_vals, label="f(x) = e^x", color='blue')  # Curva de la función
plt.scatter(x_simpson, y_simpson, color='red', label='Puntos de evaluación')  # Puntos usados en Simpson
plt.title("Regla Compuesta de Simpson")  # Título del gráfico
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("f(x)")  # Etiqueta del eje y
plt.legend()  # Mostrar la leyenda
plt.grid(True)  # Mostrar la cuadrícula
plt.show()  # Mostrar la gráfica
```
