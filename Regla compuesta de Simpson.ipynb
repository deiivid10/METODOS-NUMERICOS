```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de la funci√≥n que se va a integrar: f(x) = e^x
def f(x):
    return np.exp(x)

# Implementaci√≥n de la Regla de Simpson Compuesta (Algoritmo 4.1)
def simpson_compuesta(f, a, b, n):
    if n % 2 != 0:
        raise ValueError("n debe ser par")  # La regla de Simpson requiere que n sea par
    h = (b - a) / n  # Longitud de cada subintervalo
    x = np.linspace(a, b, n + 1)  # Se generan los n+1 puntos entre a y b
    fx = f(x)  # Se eval√∫a la funci√≥n en todos esos puntos
    # Se aplica la f√≥rmula de Simpson compuesta:
    # f(a) + f(b) + 4*f(puntos impares) + 2*f(puntos pares internos)
    S = fx[0] + fx[-1] + 4 * sum(fx[1:-1:2]) + 2 * sum(fx[2:-2:2])
    return (h / 3) * S  # Resultado final de la integral aproximada

# Par√°metros del intervalo de integraci√≥n
a = 0  # L√≠mite inferior
b = 4  # L√≠mite superior
n = 2  # N√∫mero de subintervalos (debe ser par)

# C√°lculo de la integral aproximada usando Simpson compuesta
aprox = simpson_compuesta(f, a, b, n)

# C√°lculo del valor exacto de la integral definida de e^x en [0,2]
exacto = np.exp(2)

# C√°lculo del error absoluto entre el valor exacto y el aproximado
error = abs(exacto - aprox)

# Impresi√≥n de resultados
print(f"Aproximaci√≥n con n={n}: {aprox}")
print(f"Valor exacto (exp(2)): {exacto}")
print(f"Error absoluto: {error}")

# Generaci√≥n de valores para graficar la funci√≥n original
x_vals = np.linspace(a, b, 1000)  # Muchos puntos para una curva suave
y_vals = f(x_vals)  # Evaluaci√≥n de f(x) = e^x en esos puntos

# Puntos que se usan en la Regla de Simpson
x_simpson = np.linspace(a, b, n + 1)  # Puntos de evaluaci√≥n
y_simpson = f(x_simpson)  # f(x) evaluado en esos puntos

# Crear la gr√°fica
plt.figure(figsize=(8, 5))  # Tama√±o de la figura
plt.plot(x_vals, y_vals, label="f(x) = e^x", color='blue')  # Curva de la funci√≥n
plt.scatter(x_simpson, y_simpson, color='red', label='Puntos de evaluaci√≥n')  # Puntos usados en Simpson
plt.title("Regla Compuesta de Simpson")  # T√≠tulo del gr√°fico
plt.xlabel("x")  # Etiqueta del eje x
plt.ylabel("f(x)")  # Etiqueta del eje y
plt.legend()  # Mostrar la leyenda
plt.grid(True)  # Mostrar la cuadr√≠cula
plt.show()  # Mostrar la gr√°fica
```
